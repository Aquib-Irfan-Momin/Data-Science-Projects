```python
import re

def convert_pixel(value):
    pix =str(value)

    y = [int(s) for s in re.findall(r'\d+', pix)]
    x=sorted(y)
    return x[1]
    
        
df['px_l'] = df['Resolution'].apply(convert_pixel)
```

```python
import re

def convert_pixel(value):
    pix =str(value)

    y = [int(s) for s in re.findall(r'\d+', pix)]
    x=sorted(y)
    return x[0]
    
        
df['px_w'] = df['Resolution'].apply(convert_pixel)
```

```python
def convert_os(value):
    value = str(value)
    
    if 'iOS' in value:
        return 'iOS'
    elif 'Android' in value:
        return 'Android'
        
df['Operating_System'] = df['Operating_System'].apply(convert_os)
```

```python
def convert_core(value):
    value = str(value)
    
    if 'Octa Core' in value:
        return '8'
    elif 'Hexa Core' in value:
        return '6'
    elif 'Quad Core' in value:
        return '4'
    elif 'Dual Core' in value:
        return '2'
        
df['Processor_Core'] = df['Processor_Core'].apply(convert_core)
```

## **Random Forest Regressor**

```python
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
min_max_scaler = MinMaxScaler()
X_train = min_max_scaler.fit_transform(X_train)
X_test = min_max_scaler.transform(X_test)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(X_train,y_train)
```

```python
y_pred_train = reg.predict(X_train)
y_pred_test= reg.predict(X_test)
```

```python
print("Training Accuracy:",reg.score(X_train,y_train)*100)
print("Testing Accuracy:",reg.score(X_test,y_test)*100)
```

```python
# Evaluation of Training data set

#Calculate Mean Absolute Error.
MAE = mean_absolute_error(y_train,y_pred_train)
print(f"The Mean Absolute Error (MAE) is {MAE}.")

#Calculate  Mean Squared Error
MSE = mean_squared_error(y_train,y_pred_train)
print(f"The Mean Squared Error(MSE) is {MSE}.")

#Calculate Root Mean Squared Error
RMSE = np.sqrt(MSE)
print(f"The Root Mean Squared Error(RMSE) is {RMSE}.")

#Calculate R2 Score
R2 = r2_score(y_train,y_pred_train)
print(f"The R2 Score is {R2}.")
```

```python
# Evaluation of Testing data set

#Calculate Mean Absolute Error.
MAE= mean_absolute_error(y_test,y_pred_test)
print(f"The Mean Absolute Error (MAE) is {MAE}.")

#Calculate  Mean Squared Error
MSE=mean_squared_error(y_test,y_pred_test)
print(f"The Mean Squared Error(MSE) is {MSE}.")

#Calculate Root Mean Squared Error
RMSE=np.sqrt(MSE)
print(f"The Root Mean Squared Error(RMSE) is {RMSE}.")

#Calculate R2 Score
R2=r2_score(y_test,y_pred_test)
print(f"The R2 Score is {R2}.")
```
